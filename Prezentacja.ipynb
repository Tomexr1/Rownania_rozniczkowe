## Definicja
**Nowotwór**, med. tkanka wywodząca się z prawidłowych tkanek ustroju, lecz wskutek utrwalonych cech patologicznych rozrastająca się w sposób niekontrolowany, nie poddający się czynnikom regulującym wzrost, dojrzewanie i czynności komórek, rosnący mimo wyeliminowania przyczyny, jeżeli jest ona znana [1]

## Rodzaje chorób nowotworowych

Nowotwory dzielimy na:
- nowotwory łagodne (niezłośliwe)
- nowotwory złośliwe (rak, mięsak, ...)

> ### Nowotwór łagodny cechuje:
> - powolny rozrost,
> - odgraniczenie oraz nieniszczenie tkanek otaczających,
> - brak przerzutów i nawrotów.

> Nowotwór łagodny może stać się nowotworem złośliwym, zagrażającym życiu człowieka.

> ###  Nowotwór złośliwy cechuje:
> - gwałtowane tempo wzrostu,
> - częste przerzuty oraz wznowy miejscowe.

> Nowotwór ten, w zaawansowanym stadium, może upośledzić bądź uniemożliwić funkcjonowanie kluczowych organów doprowadzając do śmierci chorego.

W dalszej części prezentacji skupimy się na rakach - najczęstszej i najbardziej znanej grupie nowotworów złośliwych.

## Przyczyny choroby

Zmiana normalnych komórek na komórki nowotworowe jest wynikiem wzajemnego oddziaływania czynników genetycznych z trzema kategoriami czynników zewnętrznych. Czynniki te, zwane kancerogennymi, mogą uszkodzić DNA prowadząc do mutacji.

> ### Kancerogeny fizyczne
> Zaliczamy do nich różnego rodzaju promieniowania.

> Z najważniejszych należy wymienić:
> - promieniowanie jonizujące,
> - promieniowanie ultrafioletowe.

> ### Kancerogeny chemiczne
>To substancje chemiczne, w przypadku których długotrawała ekspozycja prowadzi do wzrostu prawdopodobieństwa na zachorowanie na chorobę nowotworową.

> Są to między innymi:
> - azbest - materiał wykorzystywany w budownictwie,
> - składniki dymu tytoniowego,
> - alkohol,
> - aflatoksyny - jako zanieczyszczenie żywności,
> - arsen - jako zanieczyszczenie wody pitnej.

> ### Kancerogeny biologiczne 
> Są to infekcje wywołane przez niektóre wirusy, bakterie lub pasożyty. 

> Na przykład:
> - wirus brodawczaka ludzkiego,
> - wirus zapalenia wątroby typu B,
> - wirus Epsteina-Barr.

## Zapobieganie

Obecnie około 30-50% przypadkom dałoby się zapobiec poprzez unikanie czynników ryzyka i wdrażanie strategii profilaktycznych. \
Obciążenie chorobami nowotworowymi można również zmniejszyć poprzez wczesne wykrywanie raka, odpowiednie leczenie oraz opiekę nad pacjentami, u których rozwinął się nowotwór.

> ### Ryzyko zachorowania możemy zmniejszyć poprzez:
> - niezażywanie tytoniu,
> - utrzymywanie prawidłowej masy ciała,
> - stosowanie zdrowej diety,
> - regularną aktywność fizyczną,
> - ograniczenie spożycia alkoholu,
> - zaszczepienie przeciwko HPV i wirusowemu zapaleniu wątroby typu B (w przypadku należenia do grupy, dla której szczepienie jest zalecane)
> - unikanie ekspozycji na promieniowanie ultrafioletowe i stosowanie środków ochrony przeciwsłonecznej,
> - zapewnienie bezpiecznego stosowania promieniowania w opiece zdrowotnej,
> - zmniejszenie narażenia na zanieczyszczenie powietrza.

> ### Wczesne wykrywanie
> Śmiertelność z powodu raka zmniejsza się, gdy choroba jest wcześnie wykryta i leczona. Istnieją dwa elementy wczesnego wykrywania: wczesna diagnoza i badania przesiewowe.

> #### Wczesna diagnoza
> Wcześnie postawiona diagnoza zwiększa szanse na skuteczność stosowanego leczenia, tym samym poprawiając rokowania pacjenta. \
> Ważnymi jej elementami są: 
> - świadomość obajwów choroby oraz konieczności konsultacji ze specjalistą w razie ich wystąpienia,
> - dostęp do usług diagnostycznych,
> - terminowe skierowanie na leczenie.

>#### Badania przesiewowe
> Badania przesiewowe mają na celu identyfikację osób z wynikami sugerującymi określony nowotwór lub stan przednowotworowy, zanim wystąpią u nich objawy. W przypadku wykrycia nieprawidłowości podczas badań przesiewowych przeprowadza się dalsze testy w celu postawienia ostatecznej diagnozy. \
> Metoda ta nie jest skuteczna dla wszytskich rodzajów raka oraz wymaga specjalistycznego sprzętu i wykwalifikowanego personelu. Z&nbsp;tego względu wybór pacjentów do badań opiera się na wieku i wystepowaniu czynników ryzyka.

## Leczenie

Leczenie dobiera się indywidualnie, biorąc pod uwagę typ nowotworu oraz stan chorego. Głównym celem jest zazwyczaj wyleczenie raka lub znaczne przedłużenie życia pacjenta. 

> ## Metody leczenia

   > ### Leczenie systemowe
   > Polega na podawaniu leków działających ogólnoustrojowo. Wyróżniamy następujące sposoby leczenia systemowego: 
   > - ### Chemioterapia
   > Polega na podawaniu do organizmu chorego leków cytostatycznych, które zaburzają proces podziału szybko dzielących się komórek, tym samym powodując ich śmierć. Niszczone są w ten sposób komórki nowotworowe. Do jej skutków ubocznych należy oddziaływanie na zdrowe i szybko dzielące się komórki, np. komórki szpiku kostnego, przewodu pokarmowego, włosów.

<center>Zdjęcie MRI raka odbytnicy przed i po chemioterapii</center>

> - ### Hormonoterapia 
> Wykorzystywana jest w przypadku nowotworów hormonozależnych. Polega na stosowaniu leków wpływających na zmianę środowiska hormonalnego, co umożliwia hamowanie wzrostu nowotworu.
> - ### Immunoterapia
> Metoda polega na zastosowaniu leków wpływających na układ odpornościowy. Ich działanie doprowadza do nasilenia odpowiedzi immunologicznej pacjenta lub do zahamowania rozwoju raka wskutek blokowania jego czynników wzrostu.
> - ### Leczenie celowane
> Oddziaływuje przede wszystkim na komórki nowotworowe, oszczędzając większość komórek prawidłowych. Aby ją zastosować, trzeba zidentyfikować profil molekularny komórek nowotworowych i znaleźć różnice pomiędzy nimi, a komórkami prawidłowymi. Przeprowadza się analizę genetyczną komórki nowotworowej, a następnie określa się cel molekularny, jakim jest zmiana w jej budowie. W ten sposób dobiera się lek, który blokując odpowiednie receptory, pozwala zahamować rozwój choroby.

> ### Leczenie chirurgiczne
> Operacje chirurgiczne wykonywane są jako jedyna forma leczenia (np. we wczesnej fazie raka jelita grubego) lub jako element terapii skojarzonej (np. z radioterapią, chemioterapią). \
> Zabiegi przeprowadzane są z różnych wskazań: 
> - jako zabiegi radykalnego leczenia miejscowego, 
> - jako zabiegi zmniejszające masę guza w skojarzeniu z innymi sposobami leczenia (np. z chemioterapią),
> - jako zabiegi interwencyjne (np. w przypadku niedrożności przewodu pokarmowego).

> ### Chirurgia robotowa
> Rozwój technologii niesie za sobą możliwość wykorzystywania zrobotyzowanych instrumentów chirurgicznych. Urządzenia&nbsp;te&nbsp;charakteryzują się wyjątkową precyzją oraz małą inwazyjnością, co skutkuje wysoką efektywnością operacji i&nbsp;znacznym ograniczeniem ryzyka powikłań pooperacyjnych. Chirurgia robotowa sprawdza się przy wykonywaniu zabiegów m.in.&nbsp;w&nbsp;obszarze urologii, ginekologii i chirurgii ogólnej.

<center>Zdjęcie przedstawia robota chirurgicznego da Vinci znajdującego się w szpitalu Św. Anny w Piasecznie. Za&nbsp;pomocą urządzenia placówka realizuje zabiegi usunięcia guza nerki, prostaty i pęcherza moczowego.</center>

> ### Radioterapia
> Radioterapia jest sposobem miejscowego leczenia nowotworów przy wykorzystaniu energii promienistej. Do tego celu używa się urządzeń emitujących promieniowanie X (tzw. promieniowanie rentgenowskie), promieniowanie gamma, elektrony, cząstki alfa oraz promieniowanie neutronowe. \
> Ze względu na sposób napromieniowania wyróżniamy:
> - teleterapię, w której źródło promieniowania umieszczone jest w pewnej odległości od guza,
> - brachyterapię, w której źródło promieniowania umieszczone jest w bezpośredniej bliskości guza.

> ### Medycyna nuklearna
> To metoda, w której stosuje się specjalne leki (zwane radiofarmaceutykami) zawierające izotopy promieniotwórcze. Izotop emitujujący promieniowanie o dużej energii, ale o małym zasięgu wprowadzany jest do wnętrza nowotworu. W ten sposób niszczone są nieprawidłowe komórki oraz minimalizowane skutki uboczne.

> ### Opieka paliatywna
> Stosowana u nieuleczalnie chorych. Leczenie ma na celu złagodzenie, a nie wyleczenie, objawów i cierpienia spowodowanego chorobą nowotworową. Może przyczynić się do poprawy komfortu oraz jakości życia.

## Metoda jawna Eulera

```python
def euler_method(f, x0, y0, h, n):
    """
    Znajduje rozwiązanie równania y' = f(x, y) metodą jawną Eulera.
    
    :param f: funkcja prawej strony równania y' = f(x, y)
    :param x0: parametr początkowy x
    :param y0: parametr początkowy y
    :param h: krok
    :param n: liczba kroków
    :return: krotka (xs, ys) zawierająca listy kolejnych wartości x i y
    """
    x = x0
    y = y0
    xs, ys = [x0], [y0]
    for i in range(n):
        y += h * f(x, y)
        x += h
        xs.append(x)
        ys.append(y)
    return xs, ys
```

## Metoda niejawna Eulera

W metodzie niejawnej Eulera $y_n$ jest dane wzorem
\begin{equation}
    y_n = y_{n-1} + hf(t_n,y_n).
\end{equation}


Zatem w każdym kroku pętli trzeba rozwiązać równianie nieliniowe $y_n - y_{n-1} - hf(t_n,y_n) = 0$ dla zmiennej $y_n$. W celu znalezienia pierwiastka tego równania będziemy korzystać z metody Newtona (Newtona-Raphsona). Polega ona na iteratywnym znajdowaniu coraz lepszego przybliżenia pierwiastka. Algorytm zaczyna z funkcją $f$, jej pochodną $f'$ oraz pewną wartością startową $x_0$ (będziemy zaczynali z $x_0 = y_{n-1}$). Jeżeli $f$ spełnia odpowiednie warunki, to dowodzi się, że kolejne przybliżenia pierwiastka dane są wzorem rekurencyjnym
\begin{equation}
    x_{k+1} = x_k - \frac{f(x_k)}{f'(x_k)}, k = 1, 2, 3, \dotsc .
\end{equation}

```python
def backward_euler_method(f, x0, y0, h, n):
    """
    Znajduje rozwiązanie równania y' = f(x, y) metodą niejawną Eulera.
    
    :param f: funkcja prawej strony równania y' = f(x, y)
    :param x0: parametr początkowy x
    :param y0: parametr początkowy y
    :param h: krok
    :param n: liczba kroków
    :return: krotka (xs, ys) zawierająca listy kolejnych wartości x i y
    """
    x = x0
    y = y0
    xs, ys = [x0], [y0]
    for i in range(n):
        y = newton_method(lambda y: y - h * f(x, y) - ys[i], y)
        x += h
        xs.append(x)
        ys.append(y)
    return xs, ys


def newton_method(f, x0):
    """
    Znajduje miejsce zerowe funkcji f metodą Newtona.

    :param f: funkcja
    :param x0: wartość startowa
    :return: przybliżenie miejsca zerowego
    """
    x = x0
    for _ in range(100):
        x = x - f(x) / (f(x + 0.0001) - f(x)) * 0.0001
    return x
```

```python
# np 
h = 10 ** (-1)

lung_1 = lung_vol_data[lung_vol_data['ID'] == 0]
n = (lung_1['Time'].iloc[-1] - lung_1['Time'][0]) / h

def gompertz(t, y):
    return y * math.log(15) / 10

xs, ys = backward_euler_method(gompertz, lung_1['Time'][0], lung_1['Observation'][0], h, int(n))
xs2, ys2 = euler_method(gompertz, lung_1['Time'][0], lung_1['Observation'][0], h, int(n))


fig, ax = plt.subplots()
ax.plot(lung_1['Time'], lung_1['Observation'], 'bo', label='Dane')
ax.plot(xs2, ys2, 'r-', label='Euler method', alpha=0.5)
ax.plot(xs, ys, 'g-', label='Backward Euler method', alpha=0.5)
fig.legend()
```
